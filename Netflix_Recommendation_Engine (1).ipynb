```python
#pre-filtering
#now we will remove all the users that have rated less movies and
#also all those movies that has been rated less in numbers
```

```python
movie_summary=Netflix_dataset.groupby('Movie_id')['Rating'].agg(['count'])
movie_summary
```

```python
#now we will create a benchmark
benchmark=round(movie_summary['count'].quantile(0.6))
benchmark
```

```python
drop_movie=movie_summary[movie_summary['count']<benchmark].index
```

```python
len(drop_movie)
```

```python
4499-2699 # movies left are
```

```python
#now we will remove all the users that are in-active

cust_summary=Netflix_dataset.groupby('Cust_Id')['Rating'].agg(['count'])
cust_summary
```

```python
cust_benchmark=round(cust_summary['count'].quantile(0.6))
cust_benchmark
```

```python
drop_cust=cust_summary[cust_summary['count'] < cust_benchmark].index
drop_cust
```

```python
Netflix_dataset=Netflix_dataset[~Netflix_dataset['Movie_id'].isin(drop_movie)]
Netflix_dataset=Netflix_dataset[~Netflix_dataset['Cust_Id'].isin(drop_cust)]
```

```python
from surprise import Reader,Dataset ,SVD
from surprise.model_selection import cross_validate
```

```python
reader=Reader()
```

```python
#we only work with top 100K rows for quick runtime

data=Dataset.load_from_df(Netflix_dataset[['Cust_Id','Movie_id','Rating']][:100000], reader)
data
```

```python
model=SVD()  # Creating a SVD model
```

```python
cross_validate(model, data, measures=['RMSE'], cv=3)
```
